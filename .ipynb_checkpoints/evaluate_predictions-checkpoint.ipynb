```python
outputs = []
for tg in tgs:
    results = []
    
    fns = fnmatch.filter(os.listdir(input_dir),'*'+tg.replace('.csv','')+'*')
    
    for fn in fns:
        if 'DenseLoss' in fn:
            lf = fn.split('_')[-2] + fn.split('_')[-1]
        else:
            lf = fn.split('_')[-1]
        lf = lf.replace('.nc','')
        
        ds = xr.open_dataset(os.path.join(input_dir,fn))
        ds = ds.assign_coords(lf = np.array([lf]))
        if 'DenseLoss' not in lf:
            ds = ds.assign_coords(alpha = np.array([0]))
        if 'gevl' not in lf:
            ds = ds.assign_coords(gamma = np.array([0]))
        results.append(ds)
    outputs.append(xr.concat(results,dim='lf'))
output_ds = xr.concat(outputs,dim='tg')
```

```python
threshold_value = ds.y_true.isel(i=0).quantile(.98,dim='time').values
threshold_pct=98
y_true = ds.y_true.isel(i=0).values
tg = 'immingham-imm-gbr-bodc.csv'
surge_codec = xr.open_dataset('/home/jovyan/test_surge_models/CoDEC_ERA5_at_gesla3_tgs_eu_hourly_anoms.nc')
surge_codec = surge_codec.sel(time=ds.time,tg=tg)
surge_codec['surge'] = surge_codec['surge'].rolling(time=3,min_periods=1,center=True).mean()
surge_codec['surge'] = (surge_codec['surge'] - surge_codec['surge'].mean(dim='time'))/surge_codec['surge'].std(dim='time',ddof=0) #normalize
surge_codec_test=surge_codec.surge.values[np.isfinite(y_true)]
y_true = y_true[np.isfinite(y_true)]
#surge_codec_test = surge_codec.sel(tg=tg).sel(time=predictand['date'].values[idx_test][n_steps-1::]).surge.values #select test timesteps
surge_codec_test_exceedances = (surge_codec_test>=threshold_value).flatten() #find where exceeding threshold
y_true_exceedances = (y_true>=threshold_value)
print('---CoDEC---')
print('bulk correlation r='+str(np.corrcoef(surge_codec_test,y_true)[0][1]))
print('bulk RMSE='+str(rmse(surge_codec_test,y_true)))
print('Confusion matrix exceedances above {0}th percentile:'.format(threshold_pct))
print(confusion_matrix(y_true_exceedances,surge_codec_test_exceedances))

print('Correlation at timesteps where observations above {0}th percentile:'.format(threshold_pct))
print('r=' + str(np.corrcoef(surge_codec_test[y_true_exceedances],y_true[y_true_exceedances])[0][1]))
print('RMSE at timesteps where observations above {0}th percentile:'.format(threshold_pct))
print('RMSE=' + str(rmse(surge_codec_test[y_true_exceedances],y_true[y_true_exceedances])))
```
